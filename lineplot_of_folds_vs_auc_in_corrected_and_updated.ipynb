```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.__version__)
print(matplotlib.__version__)
```

```python
corrected = pd.read_excel('folds_vs_auc_in_corrected.xlsx')
```

```python
corrected.head()
```

```python
plt.figure(figsize=(8,8))
plt.xlim(0.5, 5.5, 1)
ax = sns.lineplot(x='Folds',
             y='AUC',
             hue='Model version',
             data=corrected,
             palette="Paired",
             )

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.yticks(np.arange(69.01, 75.55, 0.2))
plt.grid()
# plt.show()
plt.savefig('corrected_lineplot.png')
```
